```python
import pandas as pd
football = pd.read_csv('football.csv')
player = football[football.Nationality == 'Spain']['Wage'].value_counts(normalize = 'True',bins=4).sort_index().iloc[0]*100
player = round(player)
position= football.groupby(['Position'])['Wage'].sum()
position = position[position>2000000].index[0]
Chelsea= football[football.Club == 'Chelsea'].groupby(['Club'])['Wage'].sum()
Chelsea = Chelsea.iloc[0]
print(player, position, Chelsea)
```

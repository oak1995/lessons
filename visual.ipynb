```python
import pandas as pd
df = pd.read_csv('restaurant.csv')
df.plot()
```

```python
df['total_bill'].plot(kind = 'hist')
```

```python
df['total_bill'].plot(kind = 'hist', grid = True, title = 'шоттың жалпы сомасы')
```

```python
df[['total_bill', 'tip']].plot(kind = 'hist',
                               grid = True,
                               subplots = True,
                               title = ['шоттың жалпы сомасы', 'чаевые сомасы'],
                               legend = False)
```
